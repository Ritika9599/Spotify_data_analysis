#### Popularity of the songs

```python
popularity_trend = data.groupby('year')['popularity'].mean().reset_index()

plt.figure(figsize=(10,4))
sns.lineplot(x='year', y='popularity', data=popularity_trend, marker='o')
plt.title('Average Popularity of Songs over the years')
plt.grid(True)
```

#### Danceabilty and Energy

```python
sample_data = data.sample(1000, random_state=1)  ## here we took only sample data for better understanding....
plt.figure(figsize=(10,6))
sns.regplot(x='danceability', y='energy', data=sample_data, scatter_kws={'alpha':0.4}, line_kws={'color':'green'})
plt.title('Danceability vs Energy (Sampled with Trend Line)')
plt.xlabel('Danceability')
plt.ylabel('Energy')
```

- The above diagram shows the scatter plots with regression line or trend line. each scatter point shows the individual songs .It shows the relationship between danceability and energy. As we can see the line is positive which meansthat if the song is more danceable then it is also more energetic. we also can see that points are more scattered which means relationship is not perfectly linear like there are songs which are more danceable and less energetic. The trend line is also showing some uncertainity through shaded area.. 

```python
corr = data['danceability'].corr(data['energy'])
print("Correlation between Danceability and Energy:", round(corr, 2))
```

-- It shows that there's a weak positive relationship between danceability and energy.It means if danceability increases, energy tends to increase a little bit, but not strongly.

## Acousticness

```python
def classify_acousticness(value):      ## here we have categorized the acousticness of the songs ...
    if value < 0.3:
        return 'Electronic'
    elif value < 0.7:
        return 'Mixed'
    else:
        return 'Acoustic'
        
data['acoustic_category'] = data['acousticness'].apply(classify_acousticness)
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='acoustic_category', data=data, palette='Set1')
plt.title('Acousticness Category Distribution')
plt.xlabel('Category')
plt.ylabel('Number of Songs')
```

- It shows that songs are either Acoustic or Electronic .Very few songs are mixed.

#### Analysing Heatmap

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='PiYG')
plt.title('Correlation Heatmap')
```

- Heat map shows correlation means how strongly two numerical features are related to each other. Correlation values range from -1 to +1.
- +1 means Perfect positive relationship
- 0 mean No relationship
- -1 means Perfect negative relationship
- Color Meaning:Dark Green = Strong positive relation,Dark Pink = Strong negative relation, White/light = Weak or no relation.
- Danceability vs Valence has 0.56 means Moderately positive correlation that is as danceability increases, the song tends to sound happier.
- Acousticness vs Energy has -0.75 means Strong negative correlation, acoustic songs are usually less energetic.
- Loudness vs Energy has 0.78 means Strong positive correlation,The louder the song, the more energetic it is.
- Popularity vs Acousticness has -0.57 means Moderate negative correlation, Highly acoustic songs are less popular in general.
